```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x = np.array([6,15,24,35,47,56]).reshape((-1, 1))
y = np.array([4,20,36,45,49,38]).reshape((-1,1))
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
rscores=model.score(x,y)
print(rscores)
```

```python
print('intercept',model.intercept_)
print("slope",model.coef_)
```

```python
plt.scatter(x,y)
plt.xlabel("x features", fontsize = 20)
plt.ylabel('y labels', fontsize = 20)
plt.show()
```
